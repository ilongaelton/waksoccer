```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Import the StandardScaler module
from sklearn.preprocessing import StandardScaler

# Create an instance of StandardScaler
scaler = StandardScaler()

# Normalize the data using the StandardScaler's fit_transform method
df_market_data_scaled = scaler.fit_transform(df_market_data)

# Convert the scaled data back into a DataFrame
df_market_data_scaled = pd.DataFrame(
    df_market_data_scaled,  # Scaled data
    columns=df_market_data.columns,  # Use original column names
    index=df_market_data.index  # Use original index
)

# Display the scaled DataFrame
df_market_data_scaled.head()
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(
    df_market_data_scaled,  # Scaled data from StandardScaler
    columns=df_market_data.columns,  # Use original column names
    index=df_market_data.index  # Use original index
)

# Copy the crypto names from the original DataFrame
crypto_names = df_market_data.index

# Set the coin_id column as index
df_market_data_scaled.index = crypto_names

# Display the scaled DataFrame
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 12))

# Display the list of k-values
print(k_values)
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
for k in range(1, 12):  # Loop through k-values from 1 to 11
    # Create a KMeans model using the loop counter for the n_clusters
    kmeans = KMeans(n_clusters=k, random_state=0)
    
    # Fit the model to the data using `df_market_data_scaled`
    kmeans.fit(df_market_data_scaled)
    
    # Append the model.inertia_ to the inertia list
    inertia.append(kmeans.inertia_)

# Display the inertia values
print(inertia)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {
    "k": list(range(1, 12)),  # k-values from 1 to 11
    "inertia": inertia  # Corresponding inertia values
}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_data)

# Display the DataFrame
df_elbow.head()
```

```python
# Plot a line chart with all the inertia values computed with the different values of k
import hvplot.pandas

# Plot the Elbow curve
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xlabel="Number of Clusters (k)", 
    ylabel="Inertia", 
    width=800, 
    height=400
)
```

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
best_k = 4  # Replace with the optimal value of k determined from the Elbow curve
kmeans_model = KMeans(n_clusters=best_k, random_state=0)
```

```python
# Fit the K-Means model using the scaled DataFrame
kmeans_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
cluster_labels = kmeans_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values
print(cluster_labels)
```

```python
# Create a copy of the scaled DataFrame
df_scaled_copy = df_market_data_scaled.copy()

# Display the copy of the scaled DataFrame
df_scaled_copy.head()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_scaled_copy["Cluster"] = cluster_labels

# Display the copy of the scaled DataFrame
df_scaled_copy.head()
```

```python
# Create a scatter plot using hvPlot
df_scaled_copy.hvplot.scatter(
    x="price_change_percentage_24h", 
    y="price_change_percentage_7d", 
    by="Cluster",  # Color points by cluster labels
    hover_cols=["coin_id"],  # Add crypto names to hover information
    title="Cryptocurrency Clusters",
    xlabel="24h Price Change (%)",
    ylabel="7d Price Change (%)",
    width=800,
    height=400
)
```

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
df_market_data_pca = pca.fit_transform(df_market_data_scaled)

# Convert the PCA data to a DataFrame
df_market_data_pca = pd.DataFrame(
    df_market_data_pca,
    columns=["PC1", "PC2", "PC3"],  # Name the principal components
    index=df_market_data.index  # Retain the original index
)

# View the scaled PCA data
df_market_data_pca.head()
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component
explained_variance = pca.explained_variance_ratio_

# Print the explained variance
print(explained_variance)

# Calculate the total explained variance
total_explained_variance = explained_variance.sum()
print(f"Total Explained Variance: {total_explained_variance}")
```
